```python
import pandas as pd

# Supongamos que "consolidado" es tu DataFrame principal y "c_2022" es otro DataFrame para el 2022

# Unificar BL y BL2 (asumiendo que son iguales)
consolidado['BL'] = consolidado['BL'].combine_first(consolidado['BL2'])
consolidado.drop(columns=['BL2'], inplace=True)

# Unificar Vehículo y VIN NUMBER (donde Vehículo ya contiene "Unidad + VIN NUMBER")
consolidado['Vehículo'] = consolidado['Vehículo'].combine_first(consolidado['VIN NUMBER'])
consolidado.drop(columns=['VIN NUMBER'], inplace=True)

# Eliminar columna duplicada de Importe Transacción
consolidado.drop(columns=['Importe Transacción.1'], inplace=True)

# Eliminar Unnamed: 53 si existe
if 'Unnamed: 53' in consolidado.columns:
    consolidado.drop(columns=['Unnamed: 53'], inplace=True)

# Unificar Base de Sub BL y BL2 del 2022 en el DataFrame c_2022, si existen ambas columnas
if 'SUB BL' in c_2022.columns and 'BL 2' in c_2022.columns:
    c_2022['SUB BL'] = c_2022['SUB BL'].combine_first(c_2022['BL 2'])
    c_2022.drop(columns=['BL 2'], inplace=True)

# Unir Estado de Carga con Estado
consolidado['Estado'] = consolidado['Estado'].combine_first(consolidado['Estado de Carga'])
consolidado.drop(columns=['Estado de Carga'], inplace=True)

# Revisar qué significa Serie.1 (se imprime para análisis manual)
if 'Serie.1' in consolidado.columns:
    print("Valores únicos en Serie.1:")
    print(consolidado['Serie.1'].unique())

# ====================================================
# Procesamiento de la columna Month y la columna Año
# ====================================================

# Procesar la columna Month para separar mes y año cuando aparezcan juntos
# Se identifica si en Month aparece un guion ("-") que indique que contiene ambos datos.
mask = consolidado["Month"].astype(str).str.contains("-")
if mask.any():
    # Se separa en dos partes: la izquierda es el mes y la derecha es el año
    splits = consolidado.loc[mask, "Month"].str.split("-", expand=True)
    # Actualizar la columna Month con el valor del mes
    consolidado.loc[mask, "Month"] = splits[0]
    # Actualizar la columna Año únicamente donde se encuentre vacío
    consolidado.loc[mask & consolidado["Año"].isna(), "Año"] = splits[1]

# Función para formatear el año a 4 dígitos si aparece en formato de 2 dígitos
def format_year(x):
    try:
        x_str = str(x).strip()
        if len(x_str) == 2 and x_str.isdigit():
            return "20" + x_str
    except Exception as e:
        return x
    return x

# Aplicar el formateo a la columna Año
consolidado["Año"] = consolidado["Año"].apply(format_year)

# Unificar el formato de la columna Month: convertir abreviaturas a nombres completos en español
month_map = {
    "ene": "Enero", "jan": "Enero",
    "feb": "Febrero",
    "mar": "Marzo",
    "abr": "Abril", "apr": "Abril",
    "may": "Mayo",
    "jun": "Junio",
    "jul": "Julio",
    "ago": "Agosto", "aug": "Agosto",
    "sep": "Septiembre",
    "oct": "Octubre",
    "nov": "Noviembre",
    "dic": "Diciembre", "dec": "Diciembre"
}

# Se aplica la conversión: se pasa a minúsculas y se mapea utilizando el diccionario. Si no coincide, se mantiene el valor original.
consolidado["Month"] = consolidado["Month"].str.lower().map(month_map).fillna(consolidado["Month"])

# ====================================================
# Eliminamos cualquier columna sin datos
consolidado.dropna(axis=1, how='all', inplace=True)

# Identificar columnas únicas
columnas_unicas = consolidado.columns.tolist()

print(f"Número total de columnas en el dataset unificado: {len(columnas_unicas)}")
print("Listado de columnas en el dataset final:")
print(columnas_unicas)

# Guardar el resultado en un nuevo CSV
consolidado.to_csv('/content/gdrive/MyDrive/Datos_TAGS/Datasets/Crudo/csv/Edendred_Consolidado_Final.csv', index=False)
```

## Parte 2 - Consolidación de dataset para exploracion de datos:

Se busca eliminar las columnas menos relevantes para la exploracion de datos (Serie.1, CONCATENATE, VALID, Old BL, AU, AC)

```python
import pandas as pd
import numpy as np

# Función para limpiar y convertir columnas monetarias que puedan tener símbolos de dólar y comas
def limpiar_a_float(x):
    try:
        # Convertir a string, quitar símbolos y comas y luego a float
        if isinstance(x, str):
            x = x.replace('$','').replace(',','').strip()
        return float(x)
    except Exception:
        return np.nan

# Función para limpiar el campo de Id Conductor: quitar $ y comas, manteniendo tipo objeto (cadena)
def limpiar_id_conductor(x):
    try:
        if isinstance(x, str):
            return x.replace('$','').replace(',','').strip()
    except Exception:
        return x
    return x

# Cargar el DataFrame original
ruta_original = '/content/gdrive/MyDrive/Datos_TAGS/Datasets/Crudo/csv/Edendred_Consolidado_Limpio.csv'
final = pd.read_csv(ruta_original)

# --------------------------
# 1. Eliminar columnas innecesarias (si existen)
# --------------------------
columnas_a_eliminar = ['Serie.1', 'CONCATENATE', 'VALID', 'Old BL', 'AU', 'AC']
TAGS_USAR = final.drop(columns=[col for col in columnas_a_eliminar if col in final.columns], errors='ignore')

# --------------------------
# 2. Conversiones de tipos y limpieza de formato
# --------------------------

# a) Num Tarjeta: convertir a objeto (string)
if 'Num Tarjeta' in TAGS_USAR.columns:
    TAGS_USAR['Num Tarjeta'] = TAGS_USAR['Num Tarjeta'].astype(str)

# b) Fecha Transacción: intentar convertir a datetime
if 'Fecha Transacción' in TAGS_USAR.columns:
    TAGS_USAR['Fecha Transacción'] = pd.to_datetime(TAGS_USAR['Fecha Transacción'], errors='coerce')

# c) Hora Transacción: convertir a formato hora (si la columna solo contiene horas, se pueden combinar en un datetime o extraer la hora)
if 'Hora Transacción' in TAGS_USAR.columns:
    # Suponiendo que el formato es HH:MM:SS u otro similar, se convierte a datetime y se extrae la hora
    TAGS_USAR['Hora Transacción'] = pd.to_datetime(TAGS_USAR['Hora Transacción'], format='%H:%M:%S', errors='coerce').dt.time

# d) IEPS: convertir a float64 (se asume que puede venir con formato no numérico)
if 'IEPS' in TAGS_USAR.columns:
    TAGS_USAR['IEPS'] = TAGS_USAR['IEPS'].apply(limpiar_a_float)

# e) Importe Transacción: convertir a float64, limpiando formato ($ y comas)
if 'Importe Transacción' in TAGS_USAR.columns:
    TAGS_USAR['Importe Transacción'] = TAGS_USAR['Importe Transacción'].apply(limpiar_a_float)

# f) Comision: convertir a float64, limpiando formato si es necesario
if 'Comision' in TAGS_USAR.columns:
    TAGS_USAR['Comision'] = TAGS_USAR['Comision'].apply(limpiar_a_float)

# g) Subtotal: convertir a float64, limpiando formato
if 'Subtotal' in TAGS_USAR.columns:
    TAGS_USAR['Subtotal'] = TAGS_USAR['Subtotal'].apply(limpiar_a_float)

# h) Cantidad Mercancía: convertir a float64
if 'Cantidad Mercancía' in TAGS_USAR.columns:
    TAGS_USAR['Cantidad Mercancía'] = TAGS_USAR['Cantidad Mercancía'].apply(limpiar_a_float)

# i) KG C02: convertir a float64
if 'KG C02' in TAGS_USAR.columns:
    TAGS_USAR['KG C02'] = TAGS_USAR['KG C02'].apply(limpiar_a_float)

# j) Precio Unitario: convertir a float64
if 'Precio Unitario' in TAGS_USAR.columns:
    TAGS_USAR['Precio Unitario'] = TAGS_USAR['Precio Unitario'].apply(limpiar_a_float)

# k) Rendimiento: convertir a float64
if 'Rendimiento' in TAGS_USAR.columns:
    TAGS_USAR['Rendimiento'] = TAGS_USAR['Rendimiento'].apply(limpiar_a_float)

# l) Id Conductor: quitar $ y comas, mantener como objeto
if 'Id Conductor' in TAGS_USAR.columns:
    TAGS_USAR['Id Conductor'] = TAGS_USAR['Id Conductor'].apply(limpiar_id_conductor)

# m) Importe Neto: convertir a float64, limpiando formato ($ y comas)
if 'Importe Neto' in TAGS_USAR.columns:
    TAGS_USAR['Importe Neto'] = TAGS_USAR['Importe Neto'].apply(limpiar_a_float)

# n) IVA: convertir a float64, limpiando formato ($ y comas)
if 'IVA' in TAGS_USAR.columns:
    TAGS_USAR['IVA'] = TAGS_USAR['IVA'].apply(limpiar_a_float)

```

```python
# Recorremos cada columna del dataframe
for col in TAGS_USAR.columns:
    # Eliminamos los valores nulos para evitar confusiones con NaN
    non_null_values = TAGS_USAR[col].dropna()

    # Obtenemos el conjunto de tipos de dato presentes en la columna
    unique_types = non_null_values.apply(lambda x: type(x)).unique()

    # Si hay más de un tipo de dato, se considera que hay mezcla/inconsistencia
    if len(unique_types) > 1:
        print(f"La columna '{col}' tiene tipos de datos mixtos: {unique_types}")
```

```python
import pandas as pd
import numpy as np

# Función para redondear tiempos a intervalos de 15 minutos
def redondear_hora(tiempo_str):
    """
    Recibe un string en formato 'HH:MM:SS' y devuelve un string en formato 'HH:MM'
    redondeado al intervalo de 15 minutos más cercano:
        - El redondeo se hace de forma estándar:
          por ejemplo, si los minutos están entre 0 y 7 se redondea a HH:00;
          entre 8 y 22 a HH:15; entre 23 y 37 a HH:30; entre 38 y 52 a HH:45;
          y si es 53 o más se redondea a la siguiente hora (HH+1:00).
    """
    try:
        # Convertir el string a un objeto datetime
        dt = pd.to_datetime(tiempo_str, format='%H:%M:%S', errors='coerce')
        if pd.isna(dt):
            return np.nan
        hora = dt.hour
        minuto = dt.minute

        # Redondear los minutos al múltiplo de 15 más cercano
        # Usamos round para el redondeo; si el resultado es 60, se ajusta a la siguiente hora.
        nuevo_minuto = int(round(minuto / 15) * 15)
        if nuevo_minuto == 60:
            nuevo_minuto = 0
            hora = (hora + 1) % 24  # Ajuste en caso de que pase la medianoche

        return f"{hora:02}:{nuevo_minuto:02}"
    except Exception as e:
        # En caso de error, retorna el valor original
        return tiempo_str

# Procesar la columna "Hora Transacción": redondear la hora a intervalos de 15 minutos
if 'Hora Transacción' in TAGS_USAR.columns:
    TAGS_USAR['Hora Transacción'] = TAGS_USAR['Hora Transacción'].apply(redondear_hora)
```
